```python
#reading the data file
data = pd.read_csv("C:\Gaurav Work\Machine Learning\My Work\KNN\iris.csv")
```

```python

Min_Max = data.min()
```

```python
#lets create a figure to print of plot together
fig = plt.figure(figsize=(20,5))

#create a plot of sepal length vs sepal width for 

plt.subplot(1,3,1)              #1,3,1 means-> 1 row, 3 columns and first graph id is 1
plt.scatter(data['SepalLength'], data['SepalWidth'])
plt.title("sepal length vs sepal wisth")

#create a plot of petal length vs petal width for all
plt.subplot(1,3,2)              ##1,3,1 means-> 1 row, 3 columns and first graph id is 2
plt.scatter(data['PetalLength'], data['PetalWidth'])
plt.title("Petal length vs Petal width")

#create a sepal length vs sepal width and petal length vs petal width for all together
plt.subplot(1,3,3)
plt.scatter(data['SepalLength'], data['SepalWidth'])
plt.scatter(data['PetalLength'], data['PetalWidth'])
plt.title("length vs width")
```

```python
#create dataset for each class from main dataset
iris_setosa = data[data['Name'] == 'Iris-setosa']
iris_versicolor = data[data['Name'] == 'Iris-versicolor']
iris_virginca = data[data['Name'] == 'Iris-virginica']
```
